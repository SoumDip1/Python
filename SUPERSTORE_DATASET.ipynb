# BAR CHART

```python
##  GET THE REGION WISE SALES AND THE GRAPH OUT OF IT

aggregataion = df.groupby("Region")['Sales'].sum()
agg = aggregataion
agg
```

```python
x = agg.index
```

```python
y = agg.values
```

```python
plt.bar(x,y, color = 'green')
```

```python
# WE CAN USE MULTIPLE COLOR FOR INDIVIDUAL BARS

plt.bar(x,y, color = ['red','green','orange','blue'])
```

```python
# WE CAN USE DIRECT COLOR SET FROM THE COLOR MAP

plt.bar(x,y, color = plt.cm.viridis(0.5)) # WE NEED TO PASS A VALUE BETWEEN 0-1 INSIDE PARANTHESIS
```

```python
# WE CAN USE LINING ON GRAPH AS WELL

plt.bar(x,y, color = plt.cm.viridis(0.75), hatch='/')
plt.xlabel('Region',color = 'blue' ,labelpad = 15)
plt.ylabel('Sales', color = 'red' , labelpad = 15)
plt.title('Region VS Sales', color = 'bLack', pad = 15 )
```

# PIE CHART

```python
#  DRAW PIE CHART OUT OF DATA (PURPOSE OF THE PIE CHART IS TO HIGHLIGHT ONE SEGMENT FROM OTHERS)

labels = agg.index
values = agg.values
plt.pie(values,labels=labels, explode = [0.05,0,0,0])   # it adds space beteen the slices
```

```python
plt.pie(values,labels=labels, colors = ['red','blue', 'green', 'yellow'], explode = [0.05,0,0,0], autopct = '%1.2f%%') 
# IT IS USE FOR SHOWING THE PERCENTAGE AND '%1.2f%%' IS DENOTING THE DECIMAL POINTS
```
