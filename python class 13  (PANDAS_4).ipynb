```python

import pandas as pd

data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
    'Department': ['HR', 'Sales', 'IT', 'Sales', 'IT', 'HR', 'IT', 'Sales'],
    'Salary': [45000, 55000, 60000, 52000, 52000, 52000, 70000, 53000],
    'Age': [25, 30, 35, 40, 28, 32, 45, 38]
})
data
```

```python
data['Salary']
```

```python
data['Salary'].rank()
```

```python
# creating a column out the RANK of the salary

data['Rank'] = data['Salary'].rank(ascending=False)
```

```python
data
```

```python
# Creating column of DENSE RANK of the salaries

data['Dense Rank'] = data['Salary'].rank(method = 'dense',ascending = False)
data
```

```python
# Using LaMbda expression inside apply() function we can create another coloumn

data['Hike'] = data ['Salary'].apply(lambda x: x*1.10)
data
```

```python
# Check if the Name starts with "A" or not

data['Name Check'] = data['Name'].apply(lambda x: x.startswith("A"))
data
```

```python
# Change the last two letters of the Name into capital

data['Name Change'] = data['Name'].apply(lambda x: x[-2:].upper())
data
```

```python
# check the length of the characters of Names

data['Name length'] = data['Name'].apply(lambda x: len(x))
data
```

```python
#check if the employee falls under teen or mid age

data['Age'].apply(lambda x: "teen" if x<30 else "mid")
```

```python
#check if the employee falls under teen , mid age or senior age

data['Age'].apply(lambda x: "teen" if x<30 else "mid" if x<=40 else "senior")
```

```python
data1 =pd.DataFrame( {
    'Text': [
        ' Hello   World! ',
        'This is a SAMPLE   text.',
        'Visit   https://example.com now!!',
        'Email me at test@example.com',
        'Text preprocessing    is FUN !!!',
        'CAPITAL LETTERS    EVERYWHERE',
        'i love   python. do you?',
        '1234 is just a number',
        'Mix3d with   numb3rs & symb@ls!',
        'stopwords   are not needed sometimes'
    ]
})
data1
```

```python
# clean the data set

data1['clean_1'] = data1['Text'].apply(lambda x: x.strip())
data1
```

```python
 # convert the entire data to lowercase

data1['clean_2'] = data1['clean_1'].apply(lambda x: x.lower())
data1
```

```python
# removing all the integer form the data

data1['clean_3'] = data1['clean_2'].apply(lambda x: ''.join([char for char in x if not char.isdigit()]))
data1
```

```python
# remove all the special charactes from the data

data1['clean_4'] = data1['clean_3'].apply(lambda x: ''.join([char for char in x if char.isalpha()]))
data1
```
