# Matplotlib package as matplotlib.pyplot


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.random.randint(0,100,(10,3)),columns = ['Sales', 'Units', 'Transactions'])
data
```

```python
data['Sales']
```

```python
plt.plot(data['Sales'])
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

# CUSTOMIZATION

```python
plt.plot(data['Sales'],color= 'red')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN USE INITIALS ALSO LIKE 'B' FOR BLUE
## WE CAN USE HEX CODE 
## WE CAN USE MARKERS WHICH IS NOTHING BUT THE POINTS ON THE GRAPH
## WE CAN USE 1 TO 11 FOR MARKERS BUT 'O' IS USED AS REGULAR BASIS
```

```python
plt.plot(data['Sales'],color= '#80ff00',marker = 'o')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN CHANGE THE SIZE OF THE FIGURE 

plt.figure(figsize = (20,10))
plt.plot(data['Sales'],color= 'red', marker = 'o')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN CHANGE THE WIDTH OF THE LINE USING 'LINEWIDTH'

plt.figure(figsize = (10,5))
plt.plot(data['Sales'],color= 'b', marker = 'o', linewidth = 10)
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN CHNAGE THE SHADE OF THE COLOR AS WELL USING 'ALPHA'
## MAXIMUM SATURATION OF A COLOR IS 1 AND MINIMUM IS 0

plt.figure(figsize = (5,5))
plt.plot(data['Sales'],color= 'blue', marker = 'o',linewidth = 10, alpha = 0.25)
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN EVEN CHANGE THE EDGE COLOR OF THE MARKER AS WELL USING 'markeredgecolor'

plt.figure(figsize = (5,5))
plt.plot(data['Sales'],color= '#80ff00', marker = 'o', markeredgecolor='black')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN EVEN CHANGE THE FACE COLOR OF THE MARKER AS WELL USING 'markerfacecolor'

plt.figure(figsize = (5,5))
plt.plot(data['Sales'],color= '#80ff00', marker = 'o', markeredgecolor='black', markerfacecolor='orange')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN EVEN CHANGE THE EDGE WIDTH OF THE MARKER AS WELL USING "markeredgewidth"
 
plt.figure(figsize = (5,5))
plt.plot(data['Sales'],color= '#80ff00', marker = 'o', markeredgecolor='black',
         markerfacecolor='orange',markeredgewidth=2)
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

```python
## WE CAN EVEN CHANGE THE SIZE OF THE MARKER AS WELL USING 'markersize'

plt.figure(figsize = (5,5))
plt.plot(data['Sales'],color= '#80ff00', marker = 'o', markeredgecolor='black',
         markerfacecolor='orange',markeredgewidth=2, markersize= '10')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')
```

# DRAWING MULTIPLE PLOTS USING SUBPLOT 

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
## CHANGING THE FIGURE SIZE
## THE FORMAT IS ONE ROW TWO COLUMNS

plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
## WE CAN CHANGE THE FORMAT TO TWO RWS ONE COLOUMN AS WELL

plt.figure(figsize = (10,5))
plt.subplot(2,1,1)
plt.subplot(2,1,2)
```

```python
plt.figure(figsize = (10,3))
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
## USING 'DATA' TABLE

plt.figure(figsize = (10,5))

# PLOTTING SALES DATA

plt.subplot(2,2,1)
plt.plot(data['Sales'],color= '#80ff00', marker = 'o', markeredgecolor='black',
         markerfacecolor='orange',markeredgewidth=2, markersize= '10')
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title('Sales Distribution')

# PLOTTING UNITS DATA

plt.subplot(2,2,2)
plt.plot(data['Units'],color= '#80ff10', marker = '1', markeredgecolor='blue',
         markerfacecolor='pink',markeredgewidth=2, markersize= '10')
plt.xlabel("Index")
plt.ylabel("Units")
plt.title('Units Distribution')
```

```python
## WE CAN EVEN CUSTOMIZE THE xlabel AND ylabel
## CUSTOMIZING THE xlabel

plt.plot(data['Sales'],marker = 'o', color = 'blue')
plt.xlabel('Index', fontsize=15, color='darkgreen',labelpad=10)  # labelpad is the gap betwwn the graph and the Text
plt.ylabel('Slaes')
plt.title('Units distribution')
```

```python
## CUSTOMIZING ylabel AND title

plt.plot(data['Sales'],marker = 'o', color = 'blue')
plt.xlabel('Index', fontsize=15, color='darkgreen',labelpad=10)  
plt.ylabel('Slaes', fontsize=15, color='purple',labelpad=10)  ## labelpad is the gap betwwn the graph and the Text
plt.title('Units distribution', fontsize=20, color='orange', pad=15) ## For title we nned to use 'pad' instead of 'labelpad'
```

```python
## WE CAN USE GRID ALSO

plt.grid()
plt.plot(data['Sales'],marker = 'o', color = 'blue')
plt.xlabel('Index', fontsize=15, color='darkgreen',labelpad=10)  
plt.ylabel('Slaes', fontsize=15, color='purple',labelpad=10)  ## labelpad is the gap between the graph and the Text
plt.title('Units distribution', fontsize=20, color='orange', pad=15) ## For title we nned to use 'pad' instead of 'labelpad'
```

```python
## WE CAN CHANGE THE LINE STYLE, LINE WIDTH OF GRID ALSO

plt.grid(linestyle = ':', linewidth= 2, alpha = 0.5)
plt.plot(data['Sales'],marker = 'o', color = 'blue')
plt.xlabel('Index', fontsize=15, color='darkgreen',labelpad=10)  
plt.ylabel('Slaes', fontsize=15, color='purple',labelpad=10)  ## labelpad is the gap betwwn the graph and the Text
plt.title('Units distribution', fontsize=20, color='orange', pad=15) ## For title we nned to use 'pad' instead of 'labelpad'
```

```python
## WE CAN CUSTOMIZE THE X_TICKS (IT REFERS TO THE DIGIT OF X-AXIS)
## WE CAN CUSTOMIZE THE Y_TICKS (IT REFERS TO THE DIGIT OF Y-AXIS)

plt.plot(data['Sales'],marker = 'o', color = 'blue')
plt.xticks(fontsize = 20, rotation= 25)
plt.yticks(fontsize = 15, rotation= -25)
```

```python
## WE CAN APPLY LEGEND TO A GRAPH (IT REFERS TO A KEYWORD WHICH IS A PART OF VISUALIZATION)

plt.figure(figsize = (10,5))

plt.subplot(2,2,1)
plt.plot(data['Sales'],marker = 'o', color = 'blue', label = 'SALES')
plt.legend()      ## [label='SALES'] denotes the name of the Legend

plt.subplot(2,2,2)
plt.plot(data['Sales'],marker = 'o', color = 'green', label = 'SALES')
plt.legend(loc= 'center')      ## we cann assign the location of the legend as well
```
