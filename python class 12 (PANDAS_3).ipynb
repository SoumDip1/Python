```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['A', 'B', 'C']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Age': [23, 34, 40]})
```

```python
df1
```

```python
df2
```

```python
## INNER joins (It keeps the common elements from both the table)
## MERGE

pd.merge(df1,df2,on='ID',how='inner')
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['A', 'B', 'C']})
df2 = pd.DataFrame({'ID': [1, 2, 1], 'Age': [23, 34, 40]})
```

```python
# ONE TO MANY MAPPING using MERGE
pd.merge(df1,df2,on='ID',how='inner')
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 1], 'Name': ['A', 'B', 'C']})
df2 = pd.DataFrame({'ID': [1, 2, 1], 'Age': [23, 34, 40]})
```

```python
# MANY TO MANY MAPPING using MERGE
pd.merge(df1,df2,on='ID',how='inner')
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 1, 1], 'Name': ['A', 'B', 'C', 'D']})
df2 = pd.DataFrame({'ID': [1, 2, 1], 'Age': [23, 34, 40]})
```

```python
## ONE TO MANY mapping
pd.merge(df1,df2,on='ID',how='inner')
```

```python
import pandas as pd
import numpy as np
employees =  pd.DataFrame({
    'Emp ID' : [101,102,103,104,105,106,107],
    'Name' : ['Alice', 'Bob', 'Romeo', 'Anjan', 'David', 'Abhay', None],
    'Dept ID' : [1, 2, 1, 3, np.nan, 2, 1]
})

departments =  pd.DataFrame({
    'Dept ID' : [1, 2, 3, 4, 5, np.nan, 3],
    'DeptName' : ['HR', 'IT', 'Finance', 'Admin', 'Sales', None, 'Finance']
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on = 'Dept ID',how= 'inner')
```

```python
## LEFT JOIN

pd.merge(employees,departments ,how= 'right')
```

```python
pd.merge(employees,departments ,how= 'left')
```

```python
import pandas as pd

Customers = pd.DataFrame ({
    'Customer ID' : [1,2,2,4,5],
    'CustomerName' : ['alice', 'Bob', 'Bob', 'David', 'Eva'],
    'City' : ['Bangalore', 'Mumbai', 'Mumbai', 'Delhi', 'Chennai']
})

Orders = pd.DataFrame ({
    'Order ID' : [101,102,103,104,104],
    'Customer ID' : [1,2,3,4,4],
    'Amount' : [250,400,300,150,150]
})
```

```python
Customers
```

```python
Orders
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='inner')
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='inner').shape
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='left')
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='left').shape
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='right')
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='right').shape
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='right',indicator= True)
```

```python
pd.merge(Customers,Orders, on='Customer ID',how='outer',indicator= True)
```
