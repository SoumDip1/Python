```python
## WE HAVE MULTIPLE DATASET AVAILABLE IN SEABOR DATASET GITHUB AND WE CAN DIRECTLY IMPORT THE DATASET BY USING THE NAME

import seaborn as sns
sns.load_dataset("titanic")
```

```python
sns.histplot(df['Sales'])
```

# BOXPLOT

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.boxplot(data= df, x='Category', y= 'Sales')
plt.title('Boxplot of sales by category')
plt.show()
```

# VIOLINPLOT

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data= df, x='Category', y = 'Sales')
plt.title('Violin plot of sales by category')
plt.xticks(rotation = 30)
plt.show()
```

# COUNT PLOT

```python
plt.figure(figsize=(12, 6))
sns.countplot(data= df, x='Region')
plt.title('Count plot of orders by ')
plt.show()
```

# BARPLOT

```python
## BARPLOT OF THE TOTAL SALES

plt.figure(figsize=(12, 6))
sns.barplot(data= df, x='Ship Mode', y= 'Sales')
plt.title('Barplot of sales by Ship mode')
plt.show()
```

```python
## BARPLOT OF MEAN OF THE TOTAL SALES

plt.figure(figsize=(12, 6))
sns.barplot(data= df, x='Ship Mode', y= 'Sales', estimator = np.mean)
plt.title('Barplot of sales by Ship mode')
plt.show()
```

```python
## BARPLOT OF STANDARD DEVIATION OF THE TOTAL SALES

plt.figure(figsize=(12, 6))
sns.barplot(data= df, x='Ship Mode', y= 'Sales', estimator = np.std)
plt.title('Barplot of sales by Ship mode')
plt.show()
```

```python
## BARPLOT OF SUM OF THE TOTAL SALES

plt.figure(figsize=(12, 6))
sns.barplot(data= df, x='Ship Mode', y= 'Sales', estimator = np.sum)
plt.title('Barplot of sales by Ship mode')
plt.show()
```

# LINEPLOT

```python
plt.figure(figsize=(12, 6))
sns.barplot(data= df.sort_values('Order Date'), x='Order Date', y= 'Sales')
plt.title('Line plot of sales over Time')
plt.show()
```

# SCATTER PLOT

```python
plt.figure(figsize = (12,6))
sns.scatterplot(data= df, x= 'Sales', y= 'Profit', hue= 'Category')
plt.title('Scatter plot of Sales vs Profit by category')
plt.show()
```

# HEATMAP

```python
## HEATMAP USING THE SALES COLUMN AND RESHAPING IT TO 10 ROWS AND 10 COLOUMN FORMAT

plt.figure(figsize = (20,20))
sns.heatmap(df['Sales'].values[0:100].reshape(10,10), annot= True)
```

```python
## WE CAN CHANGE THE COLOR MAP TO DIFFERENT TYPE
## (diffetent colormaps are VIRIDIS, MAGMA, BINARY)

plt.figure(figsize = (20,20))
sns.heatmap(df['Sales'].values[0:100].reshape(10,10), annot= True, cmap= 'viridis')
```

# SWARMPLOT
BASICALLY USE FOR SMALL DATASET BECAUSE F0R A LARGE DATA SET ITS GET DIFICULT TO BE INTEPRETED

```python
sns.swarmplot(data=df.sample(1000), x= 'Category', y='Sales')
```

# REGPLOT

```python
sns.regplot(data= df, x= 'Sales', y= 'Profit')
```
